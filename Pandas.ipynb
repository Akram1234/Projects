```python
import pandas as pd
sales_data_file_name = "sales_data.xlsx"
date = pd.read_excel(sales_data_file_name, index_col=None)
date.head()
```

```python
req_data = date[['timestamp','customer id']]
req_data.head()
```

```python
cust_set = set()
tot_cust = {2010:[0 for _ in range(12)],2011:[0 for _ in range(12)]}
```

```python
for index, row in req_data.iterrows():
    ts = row["timestamp"]
    cu = row["customer id"]
    if cu not in cust_set and cu!=" ":
        year = ts.year
        month = ts.month
        tot_cust[year][month-1]+=1
        cust_set.add(cu)
        
print(tot_cust)
```

```python
cust = {}
```

```python
for index, row in req_data.iterrows():
    ts = row["timestamp"]
    cu = row["customer id"]
    y = ts.year
    m = ts.month
    if cu not in cust:
        cust[cu]=[0 for i in range(13)]
    if y==2010:
        cust[cu][0]+=1
    else:
        cust[cu][m]+=1
print(cust)
```

```python
import csv
h = ["Customer ID", "DEC", "JAN", "FEB", "MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]
with open('customer_count_in_months.csv','w', newline='') as file:
    fw = csv.writer(file)
    fw.writerow(h)
    for k,v in cust.items():
        fw.writerow([k]+v)
```

```python
new_df = date[['quantity sold','timestamp' ,'unit price','customer id']]
cust_profit={}
```

```python
for index, row in new_df.iterrows():
    ts = row["timestamp"]
    q = row["quantity sold"]
    u = row["unit price"]
    cu = row["customer id"]
    y = ts.year
    m=ts.month
    if cu not in cust_profit:
        cust_profit[cu]=[0 for _ in range(13)]
    if y==2010:
        cust_profit[cu][0]+=(q*u)
    else:
        cust_profit[cu][m]+=(q*u)

print(cust_profit)
```

```python
new_df_new_products = date[["product id", "quantity sold", "timestamp", "unit price"]]
prod_set = set()
ans={}
# {
#     product_id: [quantity_sold, unit price]
# }
```

```python
for index, row in new_df_new_products.iterrows():
    ts = row["timestamp"]
    p = row["product id"]
    if ts.year==2010 or ts.month<9:
        prod_set.add(p)
    if ts.year==2011 and ts.month>=9 and ts.month<=12 and p not in prod_set:
        if p not in ans:
            ans[p]=[row["quantity sold"], row["unit price"]]
        else:
            ans[p][0]+=row["quantity sold"]
print(ans)
# print(prod_set)
```

```python
import csv
h = ["Product Id", "Quantity Sold", "Unit Price", "Revenue Generated"]
with open('quater_new_product_revenue.csv','w', newline='') as file:
    fw = csv.writer(file)
    fw.writerow(h)
    for k,v in ans.items():
        fw.writerow([k]+v+[v[0]*v[1]])
```
